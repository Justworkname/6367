```python
img = ski.io.imread('panda.jpg')
#load
```

Sobel

```python
edges = ski.filters.sobel(img)
#filter
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()
#sub plot prep by mkaing it a list?
ax[0].imshow(img)
ax[1].imshow(edges)

fig.tight_layout()
plt.show()
#calling the list or array whichever it is
```

Sobel is a high pass filter also know as an edge detector, as we can see it works.

Low Pass

```python
gauss = ski.filters.gaussian(img,mode='mirror')
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(img)
ax[1].imshow(gauss)

fig.tight_layout()
plt.show()
#same as above
```

Gauss is a low pass filter which causses blurring and reduces edges.

```python
img_2 = ski.io.imread('cat.jpg')
#load image
```

Meijiring

```python
edges = ski.filters.meijering(img_2)
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(img_2)
ax[1].imshow(edges)

fig.tight_layout()
plt.show()
```

Mejiring is typically used on BW images and is usually usaed as a blurring filter, but it can also find edges. This is exhibited by th images above.

Laplace

```python
lap = ski.filters.laplace(img_2)
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(img_2)
ax[1].imshow(lap)

fig.tight_layout()
plt.show()
```

Laplace is a high pass filter with some obvoius issues, the kernel causes a box in each of the areas that it filters.
